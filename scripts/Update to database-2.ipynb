```python
import sqlite3
import json

# Load data from JSON file
with open("../publish/package_data.jsondb", "r") as file:
    data = json.load(file)
```

```python
columns = data["keys"]
rows = data["data"]

columns
```

```python
# Connect to SQLite database (or create it if it doesn't exist)
DANGER = False

if DANGER:
    conn = sqlite3.connect("../data/package_data.db")
    cursor = conn.cursor()


    CREATE_TABLE = """

        CREATE TABLE IF NOT EXISTS packages (
            id INTEGER PRIMARY KEY,
            name TEXT,
            description TEXT,
            mpy_version TEXT,
            pkg_version TEXT,
            publish BOOLEAN,
            stub_sources TEXT, -- json string
            path TEXT,
            hash TEXT,
            stub_hash TEXT,
            port TEXT DEFAULT "",
            board TEXT DEFAULT "",
            variant TEXT DEFAULT ""
        )
    """

    print("DANGER: Dropping table")
    DROP_TABLE = "DROP TABLE IF EXISTS packages;"
    cursor.execute(DROP_TABLE)
    conn.commit()
    # Create table
    cursor.execute(CREATE_TABLE)

    # Commit changes and close connection
    conn.commit()
    conn.close()
```

```python
import json

conn = sqlite3.connect("../data/package_data.db")
cursor = conn.cursor()

# Insert data into table
for key in iter(rows):
    package = rows[key]

    # micropython-esp32-esp32_generic-stubs
    name = package["name"]
    print(name, end=", ")
    if name.startswith("micropython-"):  # remove micropython- prefix
        name = name[12:]
    if name.endswith("-stubs"):
        name = name[:-6]
    port = name.split("-")[0]
    if "-" in name:
        board = name.split("-")[1]
    else:
        board = ""

    print(f"{port=}, {board=}")

    # print(package)
    row = f"""
    INSERT INTO packages (id, name, description, mpy_version, pkg_version, publish, stub_sources, path, hash, stub_hash, port, board)
    VALUES ({key}, '{package['name']}', '{package['description']}', '{package['mpy_version']}', '{package['pkg_version']}', {package['publish']}, '{json.dumps(package['stub_sources'])}', '{package['path']}', '{package['hash']}', '{package['stub_hash']} ', '{port}', '{board}');
    """
    # print(row)
    cursor.execute(row)
```
